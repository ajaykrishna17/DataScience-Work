Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline #plot.show() is not required if we use this line
```

```python
import numpy as np
```

```python
x= np.arange(0,10)
y= np.arange(11,21)
```

```python
#Scatter Plot
plt.scatter(x, y, c='g')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph in 2D')
```

```python
#plot

plt.plot(x,y)
```

```python
y = x*x
```

```python
plt.plot(x,y, 'r*-')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('2d diagram')
```

```python
#creating Subplots

plt.subplot(2,2,1)
plt.plot(x, y, 'g--')

plt.subplot(2, 2, 2)
plt.plot(x, y, 'b*')

plt.subplot(2, 2, 3) #two rows and columns and 3rd position
plt.plot(x, y, 'r*--')
```
